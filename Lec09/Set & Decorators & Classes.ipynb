```python
string = 'racecar'
```

```python
def palindrome_checker(word):
    n = int(len(word)/2)
    for i in range(n):
        if word[i] != word[len(word) - 1 - i]:
            return False
        
    return True
```

```python
palindrome_checker(string)
```

```python
def fibo_n_place(n):
    prev = 0
    nxt = 1
    for i in range(n):
        print(prev, end=' ')
        temp = nxt
        nxt += prev
        prev = temp
    return prev
```

```python
fibo_n_place(11)
```

```python
# Decorators
```

```python
def fibo_n_place(n):
    prev = 0
    nxt = 1
    for i in range(n):
        print(prev, end=' ')
        temp = nxt
        nxt += prev
        prev = temp
    return prev
```

```python
def memory(func):
    cache = {}
    
    def wrapper(num):
        if num in cache:
            print('No computation required')
            return cache[num]
        answer = func(num)
        cache[num] = answer
        return answer
    
    return wrapper
```

```python
fibo_n_place = memory(fibo_n_place)
```

```python
fibo_n_place(11)
```
