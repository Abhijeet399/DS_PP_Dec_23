```python
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
data, label = make_circles(n_samples = 1000, noise = 0.04, random_state = 10)
```

```python
plt.scatter(data[:, 0], data[:, 1], c = label)
```

```python
from sklearn import svm

my_svm = svm.SVC(kernel = 'poly', degree = 2)
```

```python
my_svm.fit(data, label)
```

```python
my_svm.score(data, label)
```
