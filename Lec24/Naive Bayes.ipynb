```python
label_enc = LabelEncoder()
```

```python
dataset = df.apply(label_enc.fit_transform)
```

```python
dataset = dataset.values
```

```python
data_X = dataset[:, 1:]
data_Y = dataset[:, 0]
```

```python
split = 0.9

X_train = data_X[: int(data_X.shape[0] * split), :]
Y_train = data_Y[: int(data_Y.shape[0] * split)]

X_test = data_X[int(data_X.shape[0] * split) : , :]
Y_test = data_Y[int(data_Y.shape[0] * split) : ]
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
def prior_prob(y_train, label):
    return np.sum(y_train == label)/y_train.shape[0]
```

```python
def conditional_prob(X_train, Y_train, feat_col, value, label):
    X_train = X_train[Y_train == label]
    numerator = np.sum(X_train[:, feat_col] == value)
    denominator = np.sum(Y_train == label)
    return numerator/denominator
```

```python
def predict(X_train, Y_train, X_test):   # X_test is not a 2D matrix but a row of test data
    classes = np.unique(Y_train)
    
    n_feat = X_train.shape[1]
    post_prob = []
    
    for label in classes:
        likelihood_prob = 1
        
        for cols in range(n_feat):
            likelihood_prob *= conditional_prob(X_train, Y_train, cols, X_test[cols], label)
        
        prior = prior_prob(Y_train, label)
        
        post_prob.append(likelihood_prob * prior)
    
    return np.argmax(post_prob)
```

```python
print(predict(X_train, Y_train, X_test[100]), Y_test[100])
```

```python
corr = 0

for i in range(X_test.shape[0]):
    pred = predict(X_train, Y_train, X_test[i])
    if (pred == Y_test[i]):
        corr += 1
    else:
        pass
print(corr/X_test.shape[0])
```
