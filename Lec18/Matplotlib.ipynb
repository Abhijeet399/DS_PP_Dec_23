```python
# !pip install matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
data  = np.arange(0, 11)
```

```python
data
```

```python
price_water = (2 * data) + 10
```

```python
price_water
```

```python
price_oil = data**2
```

```python
price_oil
```

```python
plt.plot(data, price_water, color = 'green', label = 'Water', linewidth=2, linestyle = '-.')
plt.plot(data, price_oil, color = 'black', label = 'Oil', linewidth=3, linestyle = '--')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Price vs Time')
plt.legend(title = 'Price:')
plt.grid()
plt.show()
```

```python
plt.scatter(data, price_water, color = 'green', label = 'Water', marker = '^')
plt.scatter(data, price_oil, color = 'black', label = 'Oil', marker = '.')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Price vs Time')
plt.legend(title = 'Price:')
plt.grid()
plt.show()
```

```python
plt.bar(['2010', '2015', '2020', '2024'], [20, 25, 37, 100], color = 'green')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Price vs Time')
```

```python
time = np.array([1, 2, 3, 4])
plt.bar(time, [10, 23, 37, 100], color = 'blue', width = 0.4)
plt.bar(time + 0.4, [20, 33, 47, 150], color = 'yellow', width = 0.4)
```
