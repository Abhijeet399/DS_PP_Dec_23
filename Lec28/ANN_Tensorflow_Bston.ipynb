```python
import tensorflow
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import boston_housing
```

```python
(train_X, train_Y), (test_X, test_Y) = boston_housing.load_data()
```

```python
print(train_X.shape, train_Y.shape, test_X.shape, test_Y.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max_scaler = MinMaxScaler()
```

```python
train_X = min_max_scaler.fit_transform(train_X)
test_X = min_max_scaler.fit_transform(test_X)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential(name = 'Kendrick_Lamar')
```

```python
model.add(Dense(200, input_dim = train_X.shape[1], activation = 'relu', name = 'Layer_1'))
model.add(Dense(150, activation = 'relu', name = 'Layer_2'))
model.add(Dense(100, activation = 'relu', name = 'Layer_3'))
model.add(Dense(50, activation = 'relu', name = 'Layer_4'))
model.add(Dense(1, activation = 'relu', name = 'Layer_5'))
```

```python
model.summary()
```

```python
import keras
model.compile(optimizer = 'Adam', loss = 'mse', metrics = ['mae'])
```

```python
hist = model.fit(train_X, train_Y, epochs = 500, batch_size = 40, validation_split = 0.1)
```

```python
plt.plot(hist.history['loss'])
```

```python
preds = model.predict(test_X)
```

```python
for i in range(len(preds)):
    print(preds[i], test_Y[i])
```
