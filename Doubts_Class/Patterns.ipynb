```python
n=3
for i in range(0,n):
    for j in range(0,n):
        if ( i==0 or j==0):
            print("*",end=" ")
        elif ( i == j and i != 0):
            print("a",end=" ")
        else:
            print("_",end=" ")
    for j in range(0,n):
        if ( i==0 or j==n-1):
            print("*",end=" ")
        elif ( i+j== n-1 and i != 0):
            print("A",end=" ")
        else:
            print("_",end=" ")
    print()
for i in range(0,n):
    for j in range(0,n):
        if ( i==n-1 or j==0):
            print("*",end=" ")
        elif ( i+j== n-1):
            print("A",end=" ")
        else:
            print("_",end=" ")
    for j in range(0,n):
        if ( i==n-1 or j==n-1):
            print("*",end=" ")
        elif ( i==j ):
            print("a",end=" ")
        else:
            print("_",end=" ")
    print()
```
